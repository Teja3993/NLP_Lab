```python
def split_text_and_clean(text):
    cleaned_chars = []
    for char in text:
        if char.isalnum() or char.isspace():
            cleaned_chars.append(char)
        else:
            # Replace other special characters with a space
            cleaned_chars.append(' ')

    cleaned_text = "".join(cleaned_chars)
    # Split the cleaned text into words and filter out empty strings
    words = [word for word in cleaned_text.split() ]
    return words

# Example usage:
example_text = "Hello, world! This is an example text with some special characters like ()*$@% and numbers 123. [brackets] and 'quotes'."
example_text = "karritej2@gmail.com"
cleaned_words = split_text_and_clean(example_text)
print(cleaned_words)
```
