```python
# Import the ngrams function from the NLTK library
from nltk.util import ngrams

def run_ngram_experiment():
    print("--- N-Gram Generation Experiment ---")

    # Method 2: Dynamic Input
    # 1. Get input string from the user
    # Note: Ensure you type a sentence with spaces (e.g., "Data Science is fun")
    gvn_str = input("Enter a string (sentence): ")

    # 2. Get the N value (Order of N-gram) from the user
    try:
        gvn_n_val = int(input("Enter the value of N (e.g., 2 for Bigram, 3 for Trigram): "))
    except ValueError:
        print("Please enter a valid integer for N.")
        return

    # 3. Tokenize the string (Split by whitespace)
    # This converts "Hello World" into ['Hello', 'World']
    splt_lst = gvn_str.split()

    # Check if we have enough words to generate N-grams
    if len(splt_lst) < gvn_n_val:
        print(f"Error: Your sentence has {len(splt_lst)} words, but you asked for {gvn_n_val}-grams.")
        print("The N value cannot be larger than the number of words.")
        return

    # 4. Generate N-Grams
    # ngrams() returns a generator, so we iterate over it or convert it to a list
    rslt_n_grms = ngrams(splt_lst, gvn_n_val)

    # 5. Output the results
    print(f"\nGenerated {gvn_n_val}-Grams:")
    for itr in rslt_n_grms:
        print(itr)

# Run the function
run_ngram_experiment()
```
