```python
import nltk
from nltk.corpus import treebank
from nltk.tag import hmm

# 1. Download ALL necessary resources
nltk.download('treebank', quiet=True)
nltk.download('universal_tagset', quiet=True)
nltk.download('punkt', quiet=True)      # Required for word tokenization
nltk.download('punkt_tab', quiet=True)  # The specific missing resource!

def run_actual_hmm():
    print("=== Training the HMM Tagger ===")

    # 2. Load pre-tagged sentences to teach the model
    # We use the Penn Treebank corpus (a famous NLP dataset)
    train_data = treebank.tagged_sents(tagset='universal')[:3000]

    # 3. Construct and train the HMM
    hmm_tagger = hmm.HiddenMarkovModelTagger.train(train_data)
    print("Training complete!\n")

    # 4. Define the input text
    txt = "The Natural Language Toolkit NLTK is a platform used for building programs"

    # 5. Tokenize and Tag
    words = nltk.word_tokenize(txt)
    tagged_output = hmm_tagger.tag(words)

    print("--- Actual HMM POS Tagging Output ---")
    print(tagged_output)

if __name__ == "__main__":
    run_actual_hmm()
```
